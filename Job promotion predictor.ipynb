```python
df = pd.get_dummies(df, columns=["city"], drop_first=True)
df
```

```python
#train test split
x = df.drop("promoted",axis = 1)
y = df["promoted"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(
    x,y,test_size=0.2,random_state= 60
)
```

```python
#import and train model
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(
    n_estimators = 100,
    random_state = 60
)
model.fit(x_train,y_train)
```

```python
#test the model
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(x_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Create a DataFrame with all expected columns from x_train, initialized to 0
new_person_df = pd.DataFrame(0, index=[0], columns=x_train.columns)

# Fill in the specific values for the new person
new_person_df["age"] = 60
new_person_df["salary"] = 60000
new_person_df["experience_years"] = 30
new_person_df["gender_encoded"] = 1# Male

# Set the city dummy variable for Rajshahi
# Assuming Rajshahi is present in x_train.columns
if 'city_Rajshahi' in new_person_df.columns:
    new_person_df["city_Rajshahi"] = 1

prediction = model.predict(new_person_df)
probability = model.predict_proba(new_person_df)

print("Promotion prediction:", prediction[0])
print("Promotion probability:", probability[0])
```
